```python
# Now drop date coloumns from our data
df.drop("Date",axis=1,inplace=True)
```

### Removing Skewness using yeo-johnson method from the respective columns

```python
skew=['LDAPS_WS','LDAPS_CC3','LDAPS_CC4','LDAPS_PPT1','LDAPS_PPT2','LDAPS_PPT3','LDAPS_PPT4','DEM','Slope']
from sklearn.preprocessing import PowerTransformer
scaler = PowerTransformer(method='yeo-johnson')

new_df[skew] = scaler.fit_transform(new_df[skew].values)
new_df[skew].head()
```

```python
new_df[skew].skew()
```

```python
# After removing skewness let's check how the data has been distributed in each column.

plt.figure(figsize=(20,20), facecolor='white')
plotnumber = 1

for column in new_df[skew]:
    if plotnumber<=9:
        ax = plt.subplot(3,3,plotnumber)
        sns.distplot(new_df[column],color='indigo',kde_kws={"shade": True},hist=False)
        plt.xlabel(column,fontsize=15)
    plotnumber+=1
plt.show()
```

```python
# Removing skewness using cube root method
new_df['LDAPS_PPT1'] = np.cbrt(new_df['LDAPS_PPT1'])
new_df['LDAPS_PPT2'] = np.cbrt(new_df['LDAPS_PPT2'])
new_df['LDAPS_PPT3'] = np.cbrt(new_df['LDAPS_PPT3'])
new_df['LDAPS_PPT4'] = np.cbrt(new_df['LDAPS_PPT4'])
```

### Model Building For 'Next_Tmax'

### Saperating features and label columns

```python
x=new_df.drop('Next_Tmax',axis=1)
y=new_df['Next_Tmax']
```

### Feature scaling by using StandardScalar

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
x
```

### Let's Build the Model1

### Spliting data into train and test sets

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=130)
```

### LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
     
```

```python
prediction = model.predict(x_test)
```

```python
#calculating error
np.mean(np.absolute(prediction-y_test))
```

```python
print('Variance score: %.2f' % model.score(x_test, y_test))
```

```python
for i in range(len(prediction)):
  prediction[i]=round(prediction[i],2)
pd.DataFrame({'Actual':y_test,'Prediction':prediction,'diff':(y_test-prediction)})
```

### Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(random_state=0)
regressor.fit(x_train,y_train)
```

```python
prediction2=regressor.predict(x_test)
np.mean(np.absolute(prediction2-y_test))
```

```python
print('Variance score: %.2f' % regressor.score(x_test, y_test))
```

```python
for i in range(len(prediction2)):
  prediction2[i]=round(prediction2[i],2)
pd.DataFrame({'Actual':y_test,'Prediction':prediction2,'diff':(y_test-prediction2)})
```

### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
regr=RandomForestRegressor(max_depth=90,random_state=0,n_estimators=100)
regr.fit(x_train,y_train)
```

```python
prediction3=regr.predict(x_test)
np.mean(np.absolute(prediction3-y_test))
```

```python
print('Variance score: %.2f' % regr.score(x_test, y_test))
```

```python
for i in range(len(prediction3)):
  prediction3[i]=round(prediction3[i],2)
pd.DataFrame({'Actual':y_test,'Prediction':prediction3,'diff':(y_test-prediction3)})
```

### We can observe that Random Forest Regression is our best fit model 

### Calculating R2-score for Random Forest Regression

```python
from sklearn.metrics import r2_score

print("Mean absolute error: %.2f" % np.mean(np.absolute(prediction3 - y_test)))
print("Residual sum of squares (MSE): %.2f" % np.mean((prediction3 - y_test) ** 2))
print("R2-score: %.2f" % r2_score(y_test,prediction3 ) )
```

### Our model achived 91% of accuracy, So let us save the model.

### Saving the model:

```python
import joblib
joblib.dump(regr,"Teperature_Forecast_Tmax.pkl")
```

### loading the model

```python
# Loading the saved model
model=joblib.load("Teperature_Forecast_Tmax.pkl")

#Prediction
prediction_Tmax = model.predict(x_test)
a = np.array(y_test)
df_final = pd.DataFrame({"Original":a,"Predicted":prediction_Tmax},index=range(len(a)))
df_final
```

### plotting the prediction

```python
plt.figure(figsize=(15,5))
sns.regplot(y_test,prediction_Tmax,color='r')
plt.xlabel('Actual ',fontsize=15)
plt.ylabel('Predicted',fontsize=15)
plt.title("Random Forest Regressor")
plt.show()
```
